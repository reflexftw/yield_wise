```python
factors = df[['Nitrogen','Phosphorus','Potassium','Rainfall','Temperature','Humidity','pH_Value']]
target=df["Crop"]
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(factors, target, test_size=0.2, shuffle=True, random_state=2)
from sklearn.preprocessing import StandardScaler, LabelEncoder


label_encoder = LabelEncoder()
new_crop_summary['Crop'] = label_encoder.fit_transform(new_crop_summary['Crop'])

sc = StandardScaler()
sc.fit(Xtrain)
Xtrain = sc.transform(Xtrain)
Xtest = sc.transform(Xtest)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
dt = DecisionTreeClassifier(criterion="entropy", random_state=2, max_depth=5)
dt.fit(Xtrain, Ytrain)
dt_pred = dt.predict(Xtest)
dt_accuracy_score = metrics.accuracy_score(Ytest, dt_pred)
print(dt_accuracy_score * 100)
dt_classification_report = metrics.classification_report(Ytest, dt_pred)
print(dt_classification_report)
```

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(Xtrain,Ytrain)
GNB_pred = GNB.predict(Xtest)
GNB_accuracy_score = metrics.accuracy_score(Ytest, GNB_pred)
print(GNB_accuracy_score)
```
